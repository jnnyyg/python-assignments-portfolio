```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# exercise A
x, y = sp.symbols('x y')
f = sp.exp(x) * sp.sin(y) + y**3
df_dx = sp.diff(f, x) # partial derivative wrt x
df_dy = sp.diff(f, y) # partial derivative wrt y
print("4a)")
print("partial derivative with respect to x:", df_dx)
print("partial derivative with respect to y:", df_dy)
print()

# exercise B
g =(x**2 * y) + (x * y**2)
dg_dx = sp.diff(g, x) # partial derivative wrt x
dg_dy = sp.diff(g, y) # partial derivative wrt y
gvx = dg_dx.subs({x: 1, y: -1})
gvy = dg_dy.subs({x: 1, y: -1})
magnitude = sp.sqrt(gvx**2 + gvy**2)
print("4b)")
print("gradient vector:", (gvx, gvy))
print("gradient vector magnitude:", magnitude)
print()

# exercise C
h = sp.ln(x**2 + y**2)
dh_dxx = sp.diff(sp.diff(h, x), x) # second partial derivative wrt x
dh_dyy = sp.diff(sp.diff(h, y), y) # second partial derivative wrt y
dh_dxy = sp.diff(sp.diff(h, y), x) # second partial derivative wrt y of the partial derivative wrt x
print("4c)")
print("second partial derivative with respect to x of partial derivative with respect to x:", dh_dxx)
print("second partial derivative with respect to y of partial derivative with respect to y:", dh_dyy)
print("second partial derivative with respect to x of partial derivative with respect to y:", dh_dxy)
print()

# exercise D
print("4d)")
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
print()

# exercise E
print("4e)")
k = x**2 + sp.sin(y)
k_func = sp.lambdify((x, y), k, 'numpy')

x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='magma') # colour change ?!?!?!!?!
plt.colorbar()
plt.title('Contour plot of $k(x, y) = x^2 + sin(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

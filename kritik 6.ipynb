```python
import numpy as np
import matplotlib.pyplot as plt

def normal_density(mean, variance, x):
    if variance > 0:
        equation = 1/(np.sqrt(2*np.pi*variance)) * np.exp((-(x-mean)**2)/(2*variance))
        return equation
    
x1 = np.linspace(-5, 5, 1000)
yvalues = normal_density(0,1,x1)
# using matplotlib to plot the normal density function for a mean value of 0 and variance^2 value of 1!

plt.plot(x1,yvalues)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Normal distrubutions")
plt.grid()
plt.show()

# calculating the average male height

def integration(mean, variance, a, b):
    changex = (b-a)/1000
    avgheight = np.sum(normal_density(mean, variance, np.linspace(a, b, 1000)) * changex)
    return avgheight

print("the average height is", integration(171,(7.1**2),162,190))
```
